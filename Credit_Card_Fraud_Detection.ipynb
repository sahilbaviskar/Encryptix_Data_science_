```python
#Import the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Counts the values of normal and fraud transactions.
# 0 - normal
# 1 - fraud
data["Class"].value_counts()
```

This data is highly unbalanced.

```python
#for fraud = 1
# otherwise 0
normal_data = data[data.Class == 0]
fraud_data = data[data.Class == 1]
```

```python
print(normal_data.shape)
print(fraud_data.shape)
```

```python
normal_data.Amount.describe()
```

```python
fraud_data.Amount.describe()
```

Mean of fraud data is too much more than the mean of normal data.
Max value is also less for it.

```python
#compare the other columns values for normal and fraud data
data.groupby("Class").mean()
```

Too much variation in the values.

Under Sampling- We take the same number of normal transactions as the fraud transactions.
normal_sample_data--->492 values out of 284315.

```python
normal_sample_data = normal_data.sample(n=492)
```

```python
#Concatenate the both data
new_data = pd.concat([normal_sample_data,fraud_data],axis=0)
```

```python
new_data.head()
```

```python
new_data["Class"].value_counts()
```

```python
new_data.groupby("Class").mean()
```

```python
#Spliting data into features and target
# x= features
# y= target
x = new_data.drop(columns="Class",axis=1)
y = new_data["Class"]
```

```python
print(x)
```

```python
print(y)
```

```python
#Spliting the data into training and testing data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

Model Training

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
#Accuracy on training data
from sklearn.metrics import accuracy_score
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction,y_train)
print("Accuracy on training data:",training_data_accuracy)
```

```python
x_test_prediction = model.predict(x_test)
print(x_test_prediction)
```

```python
y_test.head(197)
```

```python
#Accuracy on testing data
from sklearn.metrics import accuracy_score
testing_data_accuracy = accuracy_score(x_test_prediction,y_test)
print("Accuracy on testing data:",testing_data_accuracy)
```

There is too much accuracy of training data and testing data.We can predict the fraud transactions easily.
